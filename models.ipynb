## Linear Regression model

```python
#importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from math import sqrt
import numpy as np

X_train = train_df.drop("heart_rate", axis=1)
X_test = test_df.drop("heart_rate", axis=1)
y_train = train_df["heart_rate"]
y_test = test_df["heart_rate"]

# normalize the dataset
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

#building the model
model = LinearRegression()

#fitting the model on training data
model.fit(X_train, y_train)

#predicting on test data
y_pred = model.predict(X_test)

#verifying the results
print("Mean Absolute Error:", mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, y_pred)))
print('Coefficient of Determination', r2_score(y_test, y_pred))
```

## XGBoost Model

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

X_train = train_df[["respiratory data"]]
X_test = test_df[["respiratory data"]]
y_train = train_df["heart_rate"]
y_test = test_df["heart_rate"]

# normalize the dataset
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)

xgb_model = XGBRegressor(objective='reg:squarederror')

# define hyperparameters to tune
params = {
    'max_depth': [1, 2, 3, 4, 5],
    'learning_rate': [0.1, 0.2, 0.3],
    'n_estimators': [10, 20, 25, 30],
}

# set up grid search
grid_search = GridSearchCV(estimator=xgb_model, param_grid=params, cv=5, n_jobs=-1)

# fit grid search to training data
grid_search.fit(X_train, y_train)

# print best combination of hyperparameters
print(grid_search.best_params_)
```

```python
model_xgb = XGBRegressor(max_depth=grid_search.best_params_['max_depth'],
                         learning_rate=grid_search.best_params_['learning_rate'],
                         n_estimators=grid_search.best_params_['n_estimators'],
                         seed=22,
                         objective='reg:squarederror')
model_xgb.fit(X_train, y_train)

# make predictions on test set
predictions = model_xgb.predict(X_test)

print("Mean Absolute Error:", mean_absolute_error(y_test, predictions))
print("Mean Squared Error:", mean_squared_error(y_test, predictions))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, predictions)))
print('Coefficient of Determination', r2_score(y_test, predictions))
```

```python
plt.plot(y_test, label='actual')
plt.plot(predictions, label='prediction', color='red')
plt.title('Testing set')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
plt.show()
```

## LSTM model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler

train_dfl = pd.DataFrame()
train_dfl = train_dfl.append([df1, df2, df5, df7, df6, df8, df9, df10], ignore_index=True)
# val_dfl = pd.DataFrame()
# val_dfl = val_dfl.append([df1, df2], ignore_index=True)
test_dfl = pd.DataFrame()
test_dfl = test_dfl.append([df4, df3], ignore_index=True)

df_r = train_dfl.astype('float32')
X_train = df_r[["respiratory data"]]
y_train = df_r[["heart_rate"]]

# df_v = val_dfl.astype('float32')
# X_val = df_v[["respiratory data"]]
# y_val = df_v[["heart_rate"]]

df_t = test_dfl.astype('float32')
X_test = df_t[["respiratory data"]]
y_test = df_t[["heart_rate"]]

# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
X_train = scaler.fit_transform(X_train)
# X_val = scaler.fit_transform(X_val)
X_test = scaler.fit_transform(X_test)

# reshape input to be [samples, time steps, features]
X_train = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
# X_val = X_val.reshape(X_val.shape[0], 1, X_val.shape[1])
X_test = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])
# print(len(X_train), len(X_val), len(X_test))
print(len(X_train), len(X_test))

# create and fit the LSTM network
model_lstm = Sequential()
model_lstm.add(LSTM(4, batch_input_shape=(1, X_train.shape[1], X_train.shape[2])))
model_lstm.add(Dense(1))
model_lstm.compile(loss='mean_squared_error', optimizer='adam')
model_lstm.summary()
```

```python
# history = model_lstm.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=1, batch_size=64, verbose=1, shuffle=False)
history = model_lstm.fit(X_train, y_train, epochs=10, batch_size=32, verbose=1, shuffle=False)
```

```python
testPredict = model_lstm.predict(X_test, batch_size=1)
testPredict
```

```python
# calculate root mean squared error
print("Mean Absolute Error:", mean_absolute_error(y_test, testPredict))
print("Mean Squared Error:", mean_squared_error(y_test, testPredict))
print("Root Mean Squared Error:", np.sqrt(mean_squared_error(y_test, testPredict)))
print('Coefficient of Determination', r2_score(y_test, testPredict))
```
